```python
import pandas as pd
```

```python
dados = pd.read_csv('Dados/aluguel_residencial_06.csv',sep=';',index_col=0)
dados.head(3)
```

## Criando Novas Variaveis

```python
dados['ValorBruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados['Valor_m2'] = round(dados['Valor']/ dados['Area'],2)
dados['ValorTotal_m2'] = round(dados['ValorBruto']/ dados['Area'],2)
dados.head()
```

```python
casa = [ 'Casa', 'Casa de Condomínio','Casa de Vila']
```

```python
dados['Tipo_Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados.head()
```
