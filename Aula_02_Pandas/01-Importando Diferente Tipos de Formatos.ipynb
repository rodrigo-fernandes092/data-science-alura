### Obtendo dados através do link html

A pagina bloqueia acesso direto pq nao está usando um navegador. Para isso tem que fingir ter um nevegador, importando a biblioteca requests.


```python
import requests

url = 'https://www.federalreserve.gov/releases/h3/current/default.htm'

header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}


#juntamos tudo com a requests
r = requests.get(url,headers=header)


# Finalmente usamos a função read_html do pandas
df_html2 = pd.read_html(r.text)
len(df_html2)
```
